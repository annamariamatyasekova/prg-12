```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
tf_test = np.concatenate([np.random.normal])
```

Priemerna TF v case (signal):

```python
priemerna_tf = tf_test.mean(axis = 0)
plt.figure(figsize = (12,4))
plt.plot(range(60), priemerna_tf, color = "royalblue")
plt.title("Priemerna tepova frekvencia v case")
plt.xlabel("Cas (minuty)")
plt.ylabel("TF(BPM)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Boxplot - 5 fazi

```python
faze_data = [
    tf_test[:, 0:10].ravel(),
    tf_test[:, 10:20].ravel(),
    tf_test[:, 20:40].ravel(),
    tf_test[:, 40:50].ravel(),
    tf_test[:, 50:60].ravel()
]
faze_labels = ["Klud", "Zahrievanie", "Aerobne", "Sprint", "Regeneracia"]
plt.figure(figsize = (10,5))
plt.boxplot(faze_data, labels=faze_labels)
plt.title("Rozlozenie TF podla faze zatazoveho testu")
plt.xlabel("Faza")
plt.ylabel("TF(BPM")
plt.grid(True)
plt.tight_layout()
plt.show
```

Histogram - klud vs sprint

```python
klud = tf_test[:,0:10].ravel()
sprint = tf_test[:, 40:50].ravel()
plt.figure(FIGSIZE=(8,4))
plt.hist(klud, bins = 20, alpha = 0.6, color = "steelblue", label = "Klud")
plt.title("Histogram TF - Klud vs Sprint")
plt.xlabel("TF (BPM")
plt.ylabel("Pocet hodnot")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show
```
