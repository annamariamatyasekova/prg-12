```python
import numpy as np
data = np.load("pacienti.npy")
print("Tvar pola:", data.shape)
print("Prvych 5 riadkov:")
print(data[:5])
nazvy = ["Systolicky tlak", "Diastolicky tlak", "Tepova frekvencia"]
for i, nazov in enumerate(nazvy):
    print(f"{nazov}: priemer = {data[:, i].mean():.1f}, std = {data[:,i].std():.1f}")
hypertenzia = data[data[:,0]>=140]
print(f"\nPocet pacientov s hypertenziou: {len(hypertenzia)}")
print(f"\nPriemerny systolicky tlak: {hypertenzia[:,0].mean():.1f} mmHg")
pulzny_tlak = (data[:,0] - data[:, 1]).reshape(-1,1)
data_rozsirena = np.column_stack((data, pulzny_tlak))
print("\nData s pulznym tlakom (prvych 5 riadkov):")
print(data_rozsirena[:5])
index = np.argmax(data[:,2])
print(f"\nPacient s najvzssou tepovou frekvenciou: index {index}")
print(f"Cely zaznam: {data[index]}")
```
